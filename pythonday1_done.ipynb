# loc and iloc for subsetting the dataframe..

syntax for loc:
df.loc[ rowlabels, columnlabels] 

syntax for iloc:
df.iloc[rowindexes,columnindexes]

```python
res1=data2.loc[0,"BookID"]
res2=data2.iloc[0,2] #0=row_index 1=columnindex of BOOKNAME
print(res1)
res2
```

```python
data2.loc[[0,1],["GENRE","BOOKNAME"]] #subsetting multiple rows and multiple columns...
```

```python
data2.iloc[[0,1,2],[0,3]] #same output as the above but using iloc method
```

# deal with missing values

# removing rows having missing values with dropna() method

```python
df=data2.dropna(axis="rows",how="any");
df
```

# how="any" or "all"

# here any means it delete the entire row if any column contains missing value

# here "all" means it delete the entire row if and only if all the columns having missing values.

```python
df1=data2[["SERIES","USERRATINGS"]]
df1.dropna(axis="columns",how="all")
```

# combine multiple dataframes
# 1. concat
# 2. merge

# if we write axis="rows" it check the common columns
# if there are common columns it will be write once.

# if we write axis="columns" it will check the common rows
# if there are common rows it will be written once.

```python
df4=pd.concat([df2,df3],axis="columns");
df4
```

```python
df4=pd.concat([df2,df3],axis="rows");
df4
```
